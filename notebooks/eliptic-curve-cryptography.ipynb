# Eliptic Curve cryptography

$$ y^2 = x^3 + 73 $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,30, 100 )
y = np.linspace(-30, 30, 100)
X, Y = np.meshgrid(x,y)
F = X**3 - Y**2 + 73
plt.contour(X, Y, F, [0])
plt.grid(visible=True, axis='both')
plt.show()
```
